# Basic Preprocessing
# Remove tags
# lowercase
# remove stopwords

```python
import re
def remove_tags(raw_text):
  cleaned_text=re.sub(re.compile('<.*?>'),'',raw_text)
  return cleaned_text
```

```python
df['review']=df['review'].apply(remove_tags)
```

```python
df
```

```python
df['review']=df['review'].apply(lambda x:x.lower())
```

```python
df
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

sw_list = stopwords.words('english')

df['review']=df['review'].apply(lambda x:[wordlist for wordlist in x.split() if wordlist not in sw_list]).apply(lambda x:' '.join(x))
```

```python
df['review']
```

```python
X=df['review']
y=df['sentiment']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(y)
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_test
```

```python
X_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
X_train_transformed=cv.fit_transform(X_train).toarray()
X_test_transformed=cv.transform(X_test).toarray()
```

```python
X_train_transformed.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# 1. Tokenize
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X)

# 2. Convert to sequences
sequences = tokenizer.texts_to_sequences(X)
# sequences
```

```python
# 3. Padding
X_pad = pad_sequences(sequences, maxlen=100)

# 4. Model
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=64))
model.add(SimpleRNN(64))
model.add(Dense(1, activation='sigmoid'))

# 5. Compile
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 6. Train (CORRECT)
h1 = model.fit(X_pad, y, epochs=10)
```
